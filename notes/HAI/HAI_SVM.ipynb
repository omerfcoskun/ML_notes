```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df=pd.read_csv('test1.csv')
col_names = df.columns

col_names
```

```python
df['attack_P1'].value_counts()
```

```python
X=df.drop(['time','attack','attack_P1','attack_P2','attack_P3','P2_24Vdc', 'P2_ASD', 'P2_AutoGO', 'P2_CO_rpm', 'P2_Emerg',
       'P2_HILout', 'P2_MSD', 'P2_ManualGO', 'P2_OnOff', 'P2_RTR', 'P2_SIT01',
       'P2_SIT02', 'P2_TripEx', 'P2_VT01', 'P2_VTR01', 'P2_VTR02', 'P2_VTR03',
       'P2_VTR04', 'P2_VXT02', 'P2_VXT03', 'P2_VYT02', 'P2_VYT03', 'P3_FIT01',
       'P3_LCP01D', 'P3_LCV01D', 'P3_LH', 'P3_LIT01', 'P3_LL', 'P3_PIT01',
       'P4_HT_FD', 'P4_HT_LD', 'P4_HT_PO', 'P4_HT_PS', 'P4_LD', 'P4_ST_FD',
       'P4_ST_GOV', 'P4_ST_LD', 'P4_ST_PO', 'P4_ST_PS', 'P4_ST_PT01','P1_B400B', 'P1_B4022',
       'P4_ST_TT01'],axis=1)
y=df['attack_P1']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
cols = X_train.columns
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)
```

```python
X_train = pd.DataFrame(X_train, columns=[cols])
```

```python
X_test = pd.DataFrame(X_test, columns=[cols])
```

```python
X_train.describe()
```

```python
from sklearn.svm import SVC



from sklearn.metrics import accuracy_score



svc=SVC() 



svc.fit(X_train,y_train)



y_pred=svc.predict(X_test)


# compute and print accuracy score
print('Model accuracy score with default hyperparameters: {0:0.8f}'. format(accuracy_score(y_test, y_pred)))
```

```python
svc=SVC(C=100.0) 



svc.fit(X_train,y_train)



y_pred=svc.predict(X_test)



print('Model accuracy score with rbf kernel and C=100.0 : {0:0.8f}'. format(accuracy_score(y_test, y_pred)))
```

```python

y_pred_train = svc.predict(X_train)

print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train)))
```
