```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
#grafikleri satır aralarında görmek için yapılyor
%matplotlib inline
```

```python
#make_blobs veri seti import ediliyor
from sklearn.datasets import make_blobs
```

```python
X,y=make_blobs(100,2,centers=2,cluster_std=1.7)
```

```python
plt.scatter(X[:,0],X[:,1],c=y,s=50,cmap='coolwarm')
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python

model=GaussianNB()
```

```python
model.fit(X,y)
```

```python
rng=np.random.RandomState(0)
X_yeni=[-8,-10]+[33,18]*rng.rand(1000,2)
```

```python

y_yeni=model.predict(X_yeni)
```

```python
plt.scatter(X[:,0],X[:,1],c=y,s=50,cmap='coolwarm')
lim=plt.axis()
plt.scatter(X_yeni[:,0],X_yeni[:,1],c=y_yeni,s=10,cmap='coolwarm',alpha=0.3)
plt.axis(lim)
```
